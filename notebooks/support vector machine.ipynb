```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('IRIS.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
df['species']=df["species"].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
df['species']
```

```python
df0=df[df['species']==0]
df1=df[df['species']==1]
df2=df[df['species']==2]
```

```python
plt.scatter(df0['sepal_length'],df0['sepal_width'],color='red')
plt.scatter(df1['sepal_length'],df1['sepal_width'],color='blue')
plt.xlabel('sepal length')
plt.ylabel('sepal_width')
```

```python
plt.scatter(df0['petal_length'],df0['petal_width'],color='red')
plt.scatter(df1['petal_length'],df1['petal_width'],color='blue')
plt.xlabel('petal length')
plt.ylabel('petal_width')
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
X,Y
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(C=2)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
